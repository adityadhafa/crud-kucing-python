```python
# tipe data lemari, ada list sama tuple

from typing import Any, List

# List mainan kucing (bisa ditambah/dikurang)
mainan_kucing = ["bola", "tali", "kardus", "tikus mainan"]
mainan_kucing.append("bulu ayam") # Nambahin mainan baru
print(f"Mainan Acip: {mainan_kucing}")
# Hasil: Mainan Acip: ['bola', 'tali', 'kardus', 'tikus mainan', 'bulu ayam']

#kalo tuple itu intinya ya data di dalem List yang ga boleh diubah, udah gitu doang, dan ngedefinenya pake = (something)
# Tuple data lahir kucing (data pasti, gaboleh diubah)
data_lahir_acip = ("Cimahi", "20-04-2019")
# data_lahir_acip[0] = "Bandung" # <-- INI BAKAL ERROR!
print(f"Acip lahir di {data_lahir_acip[0]}")
# Hasil: Acip lahir di Cimahi


# bisa juga pake cara cpp nihh wkwk, emang gini sihh caranya
# 1. Bikin 'keranjang' kosong
list_kucing_baru: List[Any] = [] # kalo cara ngedefinnya pake kaya namaList terus: List[hintTipeDataatauAny] = [] nah itu namanya type hint

print("Masukkan 3 nama kucing baru:")
# 2. Looping 3 kali
for i in range(3):
    nama_baru = input(f"Kucing ke-{i+1}: ")
    
    # 3. Pake .append() di dalem loop
    list_kucing_baru.append(nama_baru)

print(f"Ini dia list kucing barumu: {list_kucing_baru}")

# atau kalo mau gabungin dua list itu pake .extend(listLain)
# nih misalnya ya
mainan_kucing.extend(list_kucing_baru)
```

```python
# tipe set (himpunan): <set> (intinya ga ada duplikat dan ga bisaurut)
''' 
Isinya bisa banyak, tapi:
1. Nggak ada duplikat (kalo masukin "bola" 5x, yang kesimpen cuma 1).
2. Nggak berurutan (lo nggak bisa ambil pake index [0]).
Pake kurung kurawal {} tapi nggak ada key:value
'''

# Tempat favorit Acip buat tidur (kadang di sofa, kadang di kasur)
spot_tidur_acip = {"sofa", "kasur", "kardus", "kasur", "sofa"}
print(spot_tidur_acip)
# Hasil: {'sofa', 'kardus', 'kasur'} (Urutan bisa beda, duplikat hilang)


# gunanya set itu kalo buat data besar biar cepet, dan tanpa duplikat, kaya misalnya kaya gini

kunjungan_kucing_hari_ini = ["Acip", "Cemong", "Acip", "Kempung", "Garong", "Cemong", "Acip"]

# Kita mau tau, sebenernya ada berapa kucing UNIK yang dateng?
# Pake list, lo harus looping dan if-else ribet.

# Pake set, tinggal "lempar" aja list-nya ke set:
kucing_unik = set(kunjungan_kucing_hari_ini)

print(kucing_unik)
# Hasil: {'Kempung', 'Acip', 'Garong', 'Cemong'} 
# (Urutan acak, duplikat hilang, bersih!)

# Kalo datanya 10 juta, ngecek pake list itu LAMBAT BANGET
# data_besar_list = [...] # Anggap ada 10 juta data
# "Acip" in data_besar_list # Ini bisa butuh waktu lama

# Kalo pake set, secepat kilat
data_besar_set = set(kunjungan_kucing_hari_ini)

print("Apakah Acip dateng?")
print("Acip" in data_besar_set) # Super cepat!
# Hasil: True

print("Apakah Mucil dateng?")
print("Mucil" in data_besar_set) # Super cepat!
# Hasil: False

# jadi untuk umumnya set itu dipakai ketika kita

# dan set itu ya bener bener kaya himpunan di matematika
kucing_suka_main_air = {"Acip", "Garong", "Mucil"}
kucing_suka_kardus = {"Acip", "Cemong", "Kempung", "Garong"}

# A. Irisan (&): Siapa yang suka DUA-DUANYA?
kucing_aneh = kucing_suka_main_air & kucing_suka_kardus
# atau: kucing_suka_main_air.intersection(kucing_suka_kardus)
print(f"Suka air DAN kardus: {kucing_aneh}")
# Hasil: Suka air DAN kardus: {'Acip', 'Garong'}

# B. Gabungan (|): Siapa AJA yang ada di grup?
semua_kucing = kucing_suka_main_air | kucing_suka_kardus
# atau: kucing_suka_main_air.union(kucing_suka_kardus)
print(f"Semua kucing yg diobservasi: {semua_kucing}")
# Hasil: Semua kucing yg diobservasi: {'Acip', 'Kempung', 'Cemong', 'Mucil', 'Garong'}

# C. Selisih (-): Siapa yang suka air, TAPI NGGAK suka kardus?
kucing_air_doang = kucing_suka_main_air - kucing_suka_kardus
# atau: kucing_suka_main_air.difference(kucing_suka_kardus)
print(f"Suka air doang: {kucing_air_doang}")
# Hasil: Suka air doang: {'Mucil'}


```

```python
from typing import Any

#nih misal
#type hintnya ya, list kan tipe isinya dictionary, jadi yaudah, kalo dicitionary kan isinya nanti key sama value, nah itu type hint-in
kucinge_buk_sai: list[dict[str, Any]] = [
    { # Kucing 1 (index 0)
        "nama": "Acip",
        "umur": 5,
        "mainan": ["kardus"]
    },
    { # Kucing 2 (index 1)
        "nama": "Cemong",
        "umur": 3,
        "mainan": ["bola", "tali"]
    },
    { # Kucing 3 (index 2)
        "nama": "Kempung",
        "umur": 1,
        "mainan": ["bulu ayam", "kardus", "mouse mainan"]
    }
]

#nah, gimana cara ngambil datanya, yaa kaya ngambil data di list, sekaligus kemudian nanti di dalemnya ambil lagi keynya yang mana yang mau diambil dari datanya
print(kucinge_buk_sai[0]["nama"]) # -> nahh ya gini ini lah, jadii kaya array dua dimensi gituu tapi multi variable, ciamik banget lah sebenernya ini
#ini yang keluar nanti Acip

#oke, ngambil udah, terus kalo input? gimana cara? hmm
# gimana cara nginput keee dictionary



# kalo mau ngeoutput semua kucing dan umurnya, gimana ya berarti
for kucing in kucinge_buk_sai: # ini nanti dia malah ngereturn satu dictionary dong, ngabsenin satu satu dictionary
    
    for mainan in kucing["mainan"]: # di sini kucing itu kan udah merujuk langsung ke dictionarynya, maka kucing["mainan"] langsung merujuk ke listnya mainan
        print(f"nama kucing: {kucing["nama"]} umurnya {kucing["umur"]} dan suka mainan {mainan}")
#anjayyy bener bangett, mantap mantapp
#bisa juga kalo mau ngabsenin tiap mainannya juga

# wah ini perlu banget dijelasin kaya ke anak kecil sih, perlu catetannya
# explain it to a 5 years old


for kucing in kucinge_buk_sai:
    lemnya = ", "
    list_mainan = kucing["mainan"]
    list_mainan_yang_tanpa_siku = lemnya.join(list_mainan) # jangan lupa disimpen dong hasil joinnya
    print(f"{kucing["nama"]} umurnya {kucing["umur"]} dan suka mainan {list_mainan_yang_tanpa_siku}")
    #atau bisa juga Langsung panggil .join() di dalam kurung kurawal {}
    print(f"{kucing['nama']} umurnya {kucing['umur']} dan suka mainan {', '.join(kucing['mainan'])}")    

    
```
